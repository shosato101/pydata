```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [1, 2, 3]
y2 = [3, 1, 2]
ax.plot(x, y1)
ax.plot(x, y2)

plt.show()
```

```python
x = np.arange(0.0, 15.0, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
fig, ax = plt.subplots()
ax.plot(x, y1, label="sin")
ax.plot(x, y2, label="cos")
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()
x = [1,2,3]
y = [10, 2, 3]
labels = ["spam", "ham", "egg"]
ax.barh(x, y, tick_label=labels)

plt.show()
```

```python
fig, ax = plt.subplots()

x = [1,2,3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ["spam", "ham", "egg"]
width = 0.4
ax.bar(x, y1, width=width, tick_label=labels, label="y1")

x2 = [num + width for num in x]
ax.bar(x2, y2, width=width, label="y2")

ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ["spam", "ham", "egg"]

y_total = [num1 + num2 for num1, num2 in zip(y1, y2)]
ax.bar(x, y_total, tick_label=labels, label="y1")
ax.bar(x, y2, label="y2")
ax.legend()

plt.show()
```
